```python
import requests
from bs4 import BeautifulSoup

req = requests.get('https://comic.naver.com/webtoon/list.nhn?titleId=703846&weekday=tue')
html = req.text
soup = BeautifulSoup(html, 'lxml')
```

```python
thumbnail = soup.select_one('div.thumb img')['src']
thumbnail
```

```python
for div_detail in soup.select('div.detail'):
    h2 = div_detail.select_one('h2').get_text(strip=True)
    print(h2)
    p = div_detail.select_one('p').get_text(strip=True)
    print(p)
    p_detail = div_detail.select_one('p.detail_info span').get_text(strip=True)
    p_detail_span = div_detail.select_one('span.age').get_text(strip=True)
    print(p_detail)
    print(p_detail_span)
```
