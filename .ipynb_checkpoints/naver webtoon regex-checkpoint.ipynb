```python
html = ''
with open('webtoon_list.html', 'rt') as f:
    html = f.read()
```

```python
len(html)
```

```python
import re
```

```python
sample = '''
<a href="/webtoon/list.nhn?titleId=720121&weekday=mon" class="title" title="치즈인더트랩 (재)">치즈인더트랩 (재)</a>
<a href="/webtoon/list.nhn?titleId=714185&weekday=mon" class="title" title="링크보이">링크보이</a>
'''

re.findall(r'<a.*?>([\w\s]+)', html)
```

```python
# a태그가 title이라는 class를 갖는 경우의 정규표현식 이용
re.findall(r'<a.*?class="title".*?>([\w\s]+)', html)
```

```python
# a태그 안의 링크도 가져오기
re.findall(r'<a.href="/.*?>', html)
```
